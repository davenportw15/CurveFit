```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def polynomial_row(x, degree):
    return [x ** power for power in range(0, degree + 1)]

def polynomial_matrix(points):
    degree = len(points) - 1
    rows = [polynomial_row(x, degree) for (x, _) in points]
    return np.matrix(rows)

def curve_fit(points):
    matrix = polynomial_matrix(points)
    y = [y for (_, y) in points]
    reversed_coefficients = np.linalg.solve(matrix, y)
    coefficients = reversed_coefficients[::-1]
    return np.poly1d(coefficients)
```

```python
points = [(1, 4), (2, 20), (3, -40), (5, 0), (6, 30), (7, -30), (8, 50)]
p = curve_fit(points)
x = np.linspace(0.8, 8, 1000)
y = p(x)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
ax.plot(*zip(*points), "ro")
```
